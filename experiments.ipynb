```python
# Model(name='models/embedding-gecko-001')
```

```python
documents = SimpleDirectoryReader("Data")
```

```python
doc=documents.load_data()
```

```python
print(doc[0].text)
```

load the models

```python
model=Gemini(models='gemini-1.5-pro',api_key=google_api_key)
```

```python
embedding_model=GeminiEmbedding(model_name="models/embedding-001")
```

Create a service context  --> X

```python
# #Service Context
# service_context = ServiceContext.from_defaults(
#     llm=model,
#     embed_model=embedding_model,
#     chunk_size_limit=512,
#     chunk_overlap=20
# )
```

```python
# from llama_index.core import Settings
```

```python
# Settings.llm = model
# Settings.embed_model = embedding_model
# Settings.chunk_size = 512
# Settings.chunk_overlap = 20
```

```python
index = VectorStoreIndex.from_documents(
    doc,
    llm=model,
    embed_model=embedding_model,
    chunk_size=512,
    chunk_overlap=20
)
```

```python
## Store data locally - data containing embeddings
index.storage_context.persist(persist_dir="storage")
```

```python
query_engine = index.as_query_engine() # restrict responses
```

```python
response = query_engine.query("Applications of LLMs?")
```

```python
display(Markdown(f"<b>Response:</b> {response}"))
```
